```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
plt.plot([1,2,3,4]) # matplotlib assumes this is a sequence of y values and makes x values for us. 
                    # Python ranges start from 0 and thus range is 0 - 3 (length:4) same as y.  
plt.ylabel('random numbers')
```

```python
# for ordered pairs:
plt.plot([1,2,3,4],[1, 4, 9, 16], 'ro') # makes it red circles, r* would be r stars
plt.axis([0, 6, 0, 20]) # 0 < X < 6; 0 < Y < 20
```

```python
t = np.arange(0, 10, 2) # creates a range from 0 to (end)-step. end = 10, step = 2
plt.plot(t, t, 'b*', t, t**2, 'r-', t, t**3, 'gs') # different ways to graph the np array
```

```python
# To get rid of the objects that are outputted, use semiccolon
plt.plot(t, t, 'b*', t, t**2, 'r--', t, t**3, 'gs');
```

```python
names = ['Boys', 'Girls']
values = [4, 5]
plt.figure(1, figsize = (9,3)) # 1 figure

# first subplot in the first figure
plt.subplot(131) # 1 = numrows, 3 = num columns, 1 = first object
plt.bar(names, values)

# second subplot in the first figure
plt.subplot(132)
plt.scatter(names, values)

# third subplot in the first figure
plt.subplot(133)
plt.plot(names, values)

# first subplot in escond figure
plt.figure(2)
plt.plot([4,5,6])
```

```python
# Lines also have many attributes that i can set: linewidth, dash style, antialiased...
# Use matplotlib.lines.Line2D
lines = plt.plot([1,2,3,4], )

#plt.setp(lines) # All the attributes that i can set
# plt.setp(lines)
plt.setp(lines, color='b', antialiased=False);
```

```python
ax = plt.subplot(111)
t = np.arange(0,5,0.01)
s = np.cos(2*np.pi*t)
line = plt.plot(t,s, lw=2)
plt.ylim(-2,2)
plt.annotate('local max', xy =(1,1), xytext=(3, 1.5), arrowprops=dict(facecolor='green', shrink=.1),)
```
