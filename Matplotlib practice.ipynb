```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
plt.plot([1,2,3,4]) # matplotlib assumes this is a sequence of y values and makes x values for us. 
                    # Python ranges start from 0 and thus range is 0 - 3 (length:4) same as y.  
plt.ylabel('random numbers')
```

```python
# for ordered pairs:
plt.plot([1,2,3,4],[1, 4, 9, 16], 'ro') # makes it red circles, r* would be r stars
plt.axis([0, 6, 0, 20]) # 0 < X < 6; 0 < Y < 20
```

```python
t = np.arange(0, 10, 2) # creates a range from 0 to (end)-step. end = 10, step = 2
plt.plot(t, t, 'b*', t, t**2, 'r-', t, t**3, 'gs') # different ways to graph the np array
```

```python
# To get rid of the objects that are outputted, use semiccolon
plt.plot(t, t, 'b*', t, t**2, 'r--', t, t**3, 'gs');
```

```python
names = ['Boys', 'Girls']
values = [4, 5]
plt.figure(1, figsize = (9,3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)
```
